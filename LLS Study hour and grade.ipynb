```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
from scipy import stats
```

```python
X = np.random.uniform(0, 10, 120)
Y = np.random.uniform(-3, 3, 120) + X

plt.scatter(X, Y)
```

```python
X = X.reshape((120, 1))
print(X.shape)

Y = Y.reshape((120, 1))
print(Y.shape)
```

```python
# m = (X.T X)^-1 * X.T Y
m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
m = m[0, 0]
m
```

```python
Y_pred = X * m

plt.scatter(X, Y, c ='blue')
plt.plot(X, Y_pred, c='red', lw=2)
```

```python
plt.scatter(X, Y, c='blue')
plt.plot(X, X*1.2, c='black', lw=2)
plt.plot(X, X*1.5, c='green', lw=2)
plt.plot(X, X*0.5, c='hotpink', lw=2)
```

```python
X = X.reshape((-1))
Y = Y.reshape((-1))

ans = stats.linregress(X, Y)
```

```python
plt.scatter(X, Y, c='blue')
plt.plot(X, ans.intercept + ans.slope*X, c='orange', lw=2)
```

```python
plt.scatter(X, Y, c='blue')
plt.plot(X, ans.intercept + ans.slope*X, c='orange', lw=4)
plt.plot(X, X*1.2, c='black', lw=1)
plt.plot(X, X*1.5, c='green', lw=1)
plt.plot(X, X*0.5, c='hotpink', lw=1)
plt.plot(X, Y_pred, c='red', lw=1)
```
